```python
# Filter only 'Iris-setosa' Species
sub = iris[iris['Species'] == 'Iris-setosa']

# create KDE plot
plt.figure(figsize = (8, 6))
sns.kdeplot(x = sub['SepalLengthCm'], y = sub['SepalWidthCm'], cmap = 'plasma', fill = True)

# Titles and Labels
plt.title('Iris-setosa', fontsize = 14)
plt.xlabel('SepalLengthCm', fontsize = 12)
plt.ylabel('SepalWidthCm', fontsize = 12)

plt.show()
```

```python
# Create a kde plot of sepal_length versus sepal width for setosa species of flower.
sub = iris[iris['Species'] == 'Iris-setosa']
sns.kdeplot(sub[['SepalLengthCm', 'SepalWidthCm']], shade = True, shade_lowest = False)
plt.title('Iris-setosa')
plt.xlabel('Sepal Length Cm')
plt.ylabel('Sepal Width Cm')
plt.show()
```
